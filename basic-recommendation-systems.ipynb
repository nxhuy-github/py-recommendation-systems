```python
credits = pd.read_csv("tmdb_5000_credits.csv")
movies = pd.read_csv("tmdb_5000_movies.csv")

print(credits.shape)
print(movies.shape)
```

```python
data = movies.merge(credits, left_on="id", right_on="movie_id")
data.head()
```

```python
data = data.drop(columns=['homepage', 'title_x', 'title_y', 'status','production_countries'])
data.head()
```

### Weighted Average Ratings
the formula IMDB (formerly) used to calculate weighted ratings for movies: <br>
\begin{equation*}
W =  \frac{Rv+Cm}{v+m}
\end{equation*}
where:<br>
R: average vote for the movie <br>
v: number of votes for the movie <br>
m: minimum votes required to be listed in the Top 250 <br>
C: mean vote across the dataset <br>

```python
v = data.vote_count
R = data.vote_average
C = data.vote_average.mean()
m = data.vote_count.quantile(.7)

data["weighted_average"] = (v/(v+m) * R) + (m/(v+m) * C)
```

```python
data.head()
```

### Recommender Mk1
Recommend the user a list of the highest rated movies

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

wavg = data.sort_values("weighted_average", ascending=False)

plt.figure(figsize=(16, 6))

ax = sns.barplot(x=wavg.weighted_average.head(10), y=wavg.original_title.head(10), data=wavg, palette="deep")
plt.xlim(6.75, 8.35)
plt.title('"Best" Movies by TMDB Votes', weight='bold')
plt.xlabel('Weighted Average Score', weight='bold')
plt.ylabel('Movie Title', weight='bold')
plt.show()
```

### Recommender Mk3
Recommend movies based on their weighted average ratings and their popularity scores

```python
from sklearn.preprocessing import MinMaxScaler

# Normalize the values of weighted_average and popularity
scaler = MinMaxScaler()
data_scaled = scaler.fit_transform(data[['weighted_average', 'popularity']])
data_norm = pd.DataFrame(data_scaled, columns=['weighted_average', 'popularity'])
data_norm.head()
```

```python
# 50/50 split between the scaled weighted average rating and popularity scores
data[['norm_weighted_average', 'norm_popularity']] = data_norm
data['score'] = data.norm_weighted_average*0.5 + data.norm_popularity*0.5
scored = data.sort_values("score", ascending=False)
scored[['original_title', 'norm_weighted_average', 'norm_popularity', 'score']].head(20)
```

```python
plt.figure(figsize=(16,6))

ax = sns.barplot(x=scored['score'].head(10), y=scored['original_title'].head(10), data=scored, palette='deep')

#plt.xlim(3.55, 5.25)
plt.title('Best Rated & Most Popular Blend', weight='bold')
plt.xlabel('Score', weight='bold')
plt.ylabel('Movie Title', weight='bold')
plt.show()
```

## Content-Based Filtering

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfv = TfidfVectorizer(min_df=3, max_features=None, strip_accents='unicode', analyzer='word', token_pattern=r'\w{1,}',
                     ngram_range=(1, 3), use_idf=1, smooth_idf=1, sublinear_tf=1, stop_words='english')

data.overview = data.overview.fillna('')

tfv_matrix = tfv.fit_transform(data.overview)
tfv_matrix.shape
```

#### Calculating Similarity Scores

```python
from sklearn.metrics.pairwise import sigmoid_kernel

sig = sigmoid_kernel(tfv_matrix, tfv_matrix)
indices = pd.Series(data.index, index=data['original_title']).drop_duplicates()

indices.head()
```

```python
def give_rec(title, sig=sig):
    idx = indices[title]
    sig_scores = list(enumerate(sig[idx]))
    sig_scores = sorted(sig_scores, key=lambda x: x[1], reverse=True)
    sig_scores = sig_scores[1:11]
    movie_indices = [i[0] for i in sig_scores]
    return data.original_title.iloc[movie_indices]
```

```python
# Testing our content-based recommendation system with the seminal film Spy Kids
give_rec('Spy Kids')
```
